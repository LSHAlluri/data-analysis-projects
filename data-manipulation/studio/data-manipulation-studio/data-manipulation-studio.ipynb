## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd

baltimore = pd.read_csv("baltimore_9-24-2016_9-30-2017.csv") 
boston = pd.read_csv("boston_9-24-2016_9-30-2017.csv")
new_york = pd.read_csv("new-york_9-24-2016_9-30-2017.csv")
philadelphia = pd.read_csv("philadelphia_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data to related to San Francisco. Pull up your notebook from the last lesson and use your cleaning skills to clean the dataframes as necessary.

```python
# Clean your data here!
import numpy as np
print("Percentage of missing data from each column of BALTIMORE:\n")
for col in baltimore.columns:
    pct_missing = np.mean(baltimore[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

print("\nPercentage of missing data from each column of BOSTON:\n")
for col in boston.columns:
    pct_missing = np.mean(boston[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

print("\nPercentage of missing data from each column of NEW YORK:\n")
for col in new_york.columns:
    pct_missing = np.mean(new_york[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))

print("\nPercentage of missing data from each column of PHILADELPHIA:\n")
for col in philadelphia.columns:
    pct_missing = np.mean(philadelphia[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python

# Type
# Grade - 100%
# Origin District - 100%   
# Environment - 100%

# Unit of Sale - 100%

# Quality - 100%
# Condition - 100%
# Appearance - 100%
# Storage - 100%
# Crop - 100%
# Trans Mode - 100%

print(baltimore.shape)
baltimore = baltimore.drop(['Type',"Grade",'Origin District','Environment','Quality','Condition','Appearance','Storage',
                           'Crop','Trans Mode'],
                           axis=1) 
print(baltimore.shape)

print(boston.shape)
boston = boston.drop(['Type',"Grade",'Origin District','Environment','Quality','Condition','Appearance','Storage',
                           'Crop','Trans Mode'],
                           axis=1) 
print(boston.shape)

print(new_york.shape)
new_york = new_york.drop(['Type',"Grade",'Origin District','Environment','Quality','Condition','Appearance','Storage',
                           'Crop','Trans Mode'],
                           axis=1) 
print(new_york.shape)

print(philadelphia.shape)
philadelphia = philadelphia.drop(['Type',"Grade",'Origin District','Environment','Quality','Condition','Appearance','Storage',
                           'Crop','Trans Mode'],
                           axis=1) 
print(philadelphia.shape)
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
df = pd.concat([baltimore, boston, new_york, philadelphia], axis=0, 
               keys=['baltimore', 'boston', 'new_york', 'philadelphia'])
df.head(30)
# df.tail()
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
df[["Low Price", "High Price"]].groupby(df["Unit of Sale"]).agg('mean')
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
df[[]].groupby(df["City Name", "Variety"]).agg('mean')
```
