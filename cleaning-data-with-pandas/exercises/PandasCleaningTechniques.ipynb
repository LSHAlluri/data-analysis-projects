## Missing Data

Try out different methods to locate and resolve missing data.

```python
# Try to find some missing data!
# df.isna()
#shows the total number of nulls in each column
print("Missing values in each column:\n", df.isna().sum())

# dropping rows and columns with all null values
df.dropna(how="all")
df.dropna(axis="columns", how="all")

# Replacing null values in the dataset with "Unknown"
df_filled = df.fillna('Unknown')

print(df_filled)
```

## Irregular Data

With missing data out of the way, turn your attention to any outliers. Just as we did for missing data, we first need to detect the outliers.

```python
# Keep an eye out for outliers!
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot

# Basic statistics of the dataset
print("\nBasic statistics of the dataset:::\n", df.describe())


# Histograme of Age
plot.hist(df_filled['Age'], bins=10)
plot.xlabel('Age')
plot.ylabel('Frequency')
plot.show()

# Scatter plot: Clothing Id vs Rating
plot.figure(figsize=(10, 6))
plot.scatter(df_filled['Age'], df['Rating'], color='green', edgecolor='black')
plot.title('Relationship Between Age and Rating')
plot.xlabel('Age')
plot.ylabel('Rating')
plot.grid(True)
plot.show()
```
